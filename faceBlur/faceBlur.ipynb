```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("./DATA/dog_backpack.jpg")

cv2.CascadeClassifier("./haarcascades/haarcascade_frontalface_default.xml")
```

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    ax.imshow(img,cmap=cmap)
```

```python
img_copy=img.copy()
#cv2.circle(img_copy,(500,500),150,(255,255,255),-1)
```

```python
mask_shape = (img.shape[0], img.shape[1], 1)
mask = np.full((img.shape[:2]+(1,)), 0, dtype=np.uint8)
```

```python
cv2.circle(mask,(500,500),150,(255),-1)
```

```python
blur_img=cv2.GaussianBlur(img_copy,(51,51),10)
```

```python
mask_inv = cv2.bitwise_not(mask)
bg = cv2.bitwise_and(img, img, mask=mask_inv)
face = cv2.bitwise_and(blur_img, blur_img, mask=mask)
```

```python
display(bg)
```

```python
display(face)
```

```python
combined=cv2.add(face,bg)
```
